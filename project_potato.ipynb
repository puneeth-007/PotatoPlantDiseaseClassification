```python
os.chdir(origin)
li=[]
label=[]
z=0
for i in os.listdir():
    os.chdir(origin+'\\'+i)
    current=os.getcwd()
    for j in os.listdir():
        im=cv2.imread(current+'\\'+j,0)
        im=cv2.resize(im,(100,100))
        im=im.flatten()
        li.append(im)
        label.append(i[i.index('___')+3:])
        print(z)
        z+=1

    print(i[i.index('___')+3:])
```

```python
df=pd.DataFrame(li)
```

```python
df
```

```python
df['label']=label
```

```python
df
```

```python
df.label.unique()
```

```python
df['label']=df['label'].map({'Early_blight':0,'healthy':1,'Late_blight':2})
```

```python
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('label',axis=1),df['label'],test_size=0.2)
```

```python
lr=LogisticRegression(max_iter=100000)
lr.fit(X_train,y_train)
```

```python
lry=lr.predict(X_test)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
kny=knn.predict(X_test)
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
dty=dt.predict(X_test)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
xgby=xgb.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
rfy=rf.predict(X_test)
```
